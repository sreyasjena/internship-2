```python
driver = webdriver.Chrome()
driver.get("http://www.amazon.in/")
```

```python
driver.maximize_window()
```

```python
search_product = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_product.send_keys("Mobile phone")
```

```python
search_button = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
```

```python
brand_name = []
product_name = []
price = []
return_exchange = []
availability = []
product_url = []
```

```python
for i in range(3):
    print('scrapping page', i+1)
    
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for u in url:
        product_url.append(u.get_attribute('href'))
    
    try:
        price_tags = driver.find_elements(By.XPATH,'//span[@class="a-price"]')
        for p in price_tags:
            price.append(p.text)
    except NoSuchElementException :
        price.append('-')
        
    next_button = driver.find_element(By.XPATH,"//a[ contains (text(), 'Next' ) ]")
    next_button.click()
    time.sleep(10)
```

```python
product_url
```

```python
for u in product_url:
    driver.get(u)
    time.sleep(5)
    
    try:
        product = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(product.text)
    except NoSuchElementException :
        product_name.append('-')
        
    try:
        brand = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[45]/div/table/tbody/tr[1]/td[2]/span')
        brand_name.append(brand.text)
    except NoSuchElementException :
        brand_name.append('-')
        
    try:
        rex = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]')
        return_exchange.append(rex.text)
    except NoSuchElementException :
        return_exchange.append('-')
        
    try:
        avl = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div[1]/div/div/div[2]/div/div[2]/div/form/div/div/div[4]')
        availability.append(avl.text)
    except NoSuchElementException :
        availability.append('-')
        
        
```

```python
print(len(product_url),len(product_name),len(price),len(brand_name),len(return_exchange),len(availability))
```

```python
df = pd.DataFrame({'Brand Name':brand_name,'Name Of Product':product_name,'Price':price,'Return and Exchange':return_exchange,'Availability':availability,'Url':product_url})
df
```

```python
driver = webdriver.Chrome()
driver.get("http://www.hostelworld.com/")
```

```python
driver.maximize_window()
```

```python
search_bar = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
search_bar.send_keys("London")
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button/div[2]')
search.click()
```

```python
search_button = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button')
search_button.click()
```

```python
hostel_name = []
dist_city = []
total_reviews = []
overall_reviews = []
ratings = []
Privates = []
Dorms = []
facilities = []
property_desc = []
product_url = []
```

```python
for i in range(2):
    print('scrapping page', i+1)
    
    name = driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[2]/div/div[2]/div[1]/div/div/div/div/div[5]/div/a/a/div[2]/div[1]/span')
    for n in  name:
        hostel_name.append(n.text)
        
    url = driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[2]/div/div[2]/div[1]/div/div/div/div/div[5]/div/a/a')
    for u in url:
        product_url.append(u.get_attribute('href'))
        
    dist = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
    for d in dist:
        dist_city.append(d.text)
#Scrapping privates from price

    P = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[0::2]

    for i in P:

        Privates.append(i.text.replace('\n',': ').replace('No Privates Available','NaN'))



#Scrapping Dorms from price

    P = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[1::2]

    for i in P:
        Dorms.append(i.text.replace('\n',': ').replace('No Dorms Available','NaN'))
        
    next_button = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div/div[2]/div[1]/div/div/div/div/div[6]/div/button[2]')
    time.sleep(5)
```

```python
product_url
```

```python
for u in product_url:
    driver.get(u)
    time.sleep(1)
    
    try:
        desc = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[2]/div[4]/div/div[2]')
        property_desc.append(desc.text)
    except NoSuchElementException:
        property_desc.append('-')
        
    try:
        review_tot = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[1]/div[1]/div[2]/div[2]')
        total_reviews.append(review_tot.text)
    except NoSuchElementException:
        total_reviews.append('-')
        
    try:
        review_ov = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[1]/div[1]/div[2]/div[1]')
        overall_reviews.append(review_ov.text)
    except NoSuchElementException:
        overall_reviews.append('-')
        
    try:
        rating = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[1]/div[1]/div[1]')
        ratings.append(rating.text)
    except NoSuchElementException:
        ratings.append('-')
        
    try:
        facility = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[13]/div/ul/li[2]/ul')
        facilities.append(facility.text)
    except NoSuchElementException:
        facilities.append('-')
```

```python
df = pd.DataFrame({'Hostel Name':hostel_name,'Distance from city':dist_city,'Ratings':ratings,'Total review':total_reviews,'Overall review':overall_reviews,'Property Description':property_desc,'Facilities':facilities})
df
```
