# Question 1 & 2)

```python
driver = webdriver.Chrome()
driver.get("http://www.amazon.in/")
```

```python
driver.maximize_window()
```

```python
search_product = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_product.send_keys("Mobile phone")
```

```python
search_button = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
```

```python
brand_name = []
product_name = []
price = []
return_exchange = []
availability = []
product_url = []
```

```python
for i in range(3):
    print('scrapping page', i+1)
    
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for u in url:
        product_url.append(u.get_attribute('href'))
    
    try:
        price_tags = driver.find_elements(By.XPATH,'//span[@class="a-price"]')
        for p in price_tags:
            price.append(p.text)
    except NoSuchElementException :
        price.append('-')
        
    next_button = driver.find_element(By.XPATH,"//a[ contains (text(), 'Next' ) ]")
    next_button.click()
    time.sleep(10)
```

```python
product_url
```

```python
for u in product_url:
    driver.get(u)
    time.sleep(5)
    
    try:
        product = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(product.text)
    except NoSuchElementException :
        product_name.append('-')
        
    try:
        brand = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[45]/div/table/tbody/tr[1]/td[2]/span')
        brand_name.append(brand.text)
    except NoSuchElementException :
        brand_name.append('-')
        
    try:
        rex = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]')
        return_exchange.append(rex.text)
    except NoSuchElementException :
        return_exchange.append('-')
        
    try:
        avl = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div[1]/div/div/div[2]/div/div[2]/div/form/div/div/div[4]')
        availability.append(avl.text)
    except NoSuchElementException :
        availability.append('-')
        
        
```

```python
print(len(product_url),len(product_name),len(price),len(brand_name),len(return_exchange),len(availability))
```

```python
df = pd.DataFrame({'Brand Name':brand_name,'Name Of Product':product_name,'Price':price,'Return and Exchange':return_exchange,'Availability':availability,'Url':product_url})
df
```

```python
world_billionaires = driver.find_element(By.XPATH,'/html/body/div[1]/header/nav/div[1]/div[1]/div/div[2]/ul/li[2]/div[2]/div[3]/ul/li[1]')
world_billionaires.click()
```

```python
rank =[]
name = []
net_worth = []
age = []
citizenship = []
source = []
industry = []
```

```python
for i in range(14):
    print('scrapping page', i+1)
    
    rank_r = driver.find_elements(By.XPATH,'//div[1][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for r in rank_r:
        rank.append(r.text)
        
    name_r = driver.find_elements(By.XPATH,'//div[2][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for n in name_r:
        name.append(n.text)
        
    net_worth_r = driver.find_elements(By.XPATH,'//div[3][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for i in net_worth_r:
        net_worth.append(i.text)
        
    age_r = driver.find_elements(By.XPATH,'//div[4][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for a in age_r:
        age.append(a.text)
        
    citizenship_r = driver.find_elements(By.XPATH,'//div[5][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for c in citizenship_r:
        citizenship.append(c.text)
        
    source_r = driver.find_elements(By.XPATH,'//div[6][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for s in source_r:
        source.append(s.text)
        
    industry_r = driver.find_elements(By.XPATH,'//div[7][@class="TableRow_cell__db-hv Table_cell__houv9"]')
    for i in industry_r:
        industry.append(i.text)
        
    try:    
        next_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]/div/span[2]')
        next_button.click()
    except ElementClickInterceptedException:
        break
    time.sleep(5)
    
    
    
```

```python
print(len(rank),len(name),len(net_worth),len(age),len(citizenship),len(source),len(industry))
```

```python
df = pd.DataFrame({'Rank':rank,'Name':name,'Net Worth':net_worth,'Age':age,'Citizenship':citizenship,'Source':source,'Industry':industry})
df
```
