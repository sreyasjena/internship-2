# Question 1)

```python
driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
driver.maximize_window()
```

```python
rank = []
name = []
artist = []
upload_date = []
views =[]
```

```python
rank_tag = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[1]')
for r in rank_tag:
    rank.append(r.text)
    
name_tag = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[2]')
for n in name_tag:
    name.append(n.text)
name = [item.split('[')[0].strip() for item in name]
    
    
artist_tag = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[3]')
for a in artist_tag:
    artist.append(a.text)
    
ud = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[5]')
for u in ud:
    upload_date.append(u.text)
    
vw = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr/td[4]')
for v in vw:
    views.append(v.text)
```

```python
print(len(rank),len(name),len(artist),len(upload_date),len(views))
```

```python
df = pd.DataFrame({'RANK':rank,'NAME OF VIDEO':name,'ARTIST':artist,'VIEWS (IN BILLION)':views,'UPLOADED ON':upload_date})
df
```

# Question 8)

```python
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
driver.maximize_window()
```

```python
datasets = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
datasets.click()
```

```python
expand = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
expand.click()
```

```python
name = []
data_type = []
task = []
attribute_type = []
instances_n = []
attribute_n = []
year = []
url = []
```

```python
for i in range(10):
    print('scrapping page', i+1)
    
    Name = driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
    for n in Name:
        name.append(n.text)
    
    Year = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[3]')
    for y in Year:
        year.append(y.text)
    year = [Year.split('/')[-1] for Year in year]
    
    i_n = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[3]/span')
    for i in i_n:
        instances_n.append(i.text)
    
    a_n = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[4]/span')
    for a in a_n:
        attribute_n.append(a.text)
    
    URL = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/h2/a')
    for u in URL:
        url.append(u.get_attribute('href'))
    
    a_type = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[2]')
    for a in a_type:
        attribute_type.append(a.text)
    

    next_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    
    time.sleep(5)

```

```python
print(len(url),len(year),len(instances_n),len(attribute_n),len(attribute_type))
```

```python
url
```

```python
for u in url:
    driver.get(u)
    time.sleep(5)
    
    try:
        Task = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        task.append(Task.text)
        
    except NoSuchElementException:
        task.append('-')
        
    try:
        dt = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        data_type.append(dt.text)
        
    except NoSuchElementException:
        data_type.append('-')
```

```python
df = pd.DataFrame({'Dataset name':name,' Data type':data_type,'Task':task,'Attribute type':attribute_type,'No. of instances':instances_n,'No. of attribute':attribute_n,'Year':year})
df
```
